```python
import yfinance as yf
import pandas as pd
```

```python
def get_options(ticker):
    # Call the symbol/ticker
    yf_ticker = yf.Ticker(ticker)
    
    # Get the current stock price
    current_price = yf_ticker.history(period='1d')['Close'][0]
    
    # Create an empty list
    calls_bucket = []
    
    # Iterate through dataframes for each expiration date
    for i in range(0, len(yf_ticker.options)):
        option = yf_ticker.option_chain(yf_ticker.options[i])
        
        # Append each dataframe to the list
        calls_bucket.append(option.calls)
        
    # Concatenate dataframes 
    df_calls = pd.concat(calls_bucket)    
    
    # Add the stock price to the dataframe
    df_calls['StockPrice'] = current_price
    df_calls = df_calls[['contractSymbol', 'strike', 'lastPrice', 'StockPrice']]
    
    return df_calls
```

```python
# Read the csv file into a pandas DataFrame
df = pd.read_csv("tickers.csv", header=0, names=['Ticker', 'Name', 'Exchange'])

# Extract the tickers column as a list
ticker_list = df['Ticker'].tolist()

print(ticker_list)
```

```python
# Use slicing to get the first 50 tickers
tickers_to_test = ticker_list[:200]
```

```python
# Create an empty list to store the DataFrames
options_bucket = []

# Loop through the list of tickers
for ticker in ticker_list:
    try:
        # Call the get_options function for each ticker
        options = get_options(ticker)
        
        # Append the returned DataFrame to the options_bucket list
        options_bucket.append(options)
    except:
        # If an error occurs, print a message and move on to the next ticker
        print(f"Error fetching data for {ticker}. Moving on to the next ticker.")

# Concatenate all DataFrames in the options_bucket list into one DataFrame
all_options = pd.concat(options_bucket)
all_options.head()
```

```python
#filter by expiration date
def filter_by_string(df, string=None):
    if string is not None:
        df = df[df['contractSymbol'].str.contains(string)]
    return df

all_options_filtered = filter_by_string(all_options, '230217')
all_options_filtered.head()
```

```python
# Create a new dataframe with only the required columns
df_cleaned = all_options_filtered[['contractSymbol','strike', 'lastPrice', 'StockPrice']]

# Calculate the 3% threshold for strike
strike_threshold = df_cleaned['StockPrice'] * 1.02

# Calculate the 10% threshold for lastPrice
lastPrice_threshold = df_cleaned['StockPrice'] * 0.05

# Filter the data based on the strike and lastPrice criteria
df_cleaned = df_cleaned[(df_cleaned['strike'] >= strike_threshold) & (df_cleaned['lastPrice'] >= lastPrice_threshold)]

# Reset the index of the cleaned dataframe
df_cleaned = df_cleaned.reset_index(drop=True)

# Print the head of the cleaned dataframe
print(df_cleaned.head())
```
